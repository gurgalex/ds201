```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_pop = pd.read_excel("state_population.xlsx")
df_uni_ranks = pd.read_csv("National Universities Rankings.csv", encoding="latin1")
df_state_abrev = pd.read_csv("states_abrev.csv")
df_cohorts = pd.read_csv("Most-Recent-Cohorts-All-Data-Elements.csv")
```

```python
df_state_abrev = df_state_abrev.drop(8)
```

```python
df_pop.head()
```

```python
df_state_abrev = df_state_abrev.reindex(df_pop.index)
```

```python
mask_non_states = np.logical_not(df_pop["State or territory"].
                      isin(df_state_abrev["State"]))
```

```python
df_pop = df_pop.drop(df_pop[mask_non_states].index)
```

```python
df_pop.reset_index(drop=True, inplace=True)
```

```python
df_pop = df_pop.iloc[:,[0,2,3]]
```

```python
df_pop = df_pop.rename(columns = {"State or territory": "State",
                                  "Rank in the fifty states, 2016": "Rank by Population",
                                  "Population estimate, July 1, 2016[4]": "Population",
                                 })
```

```python
df_state_abrev = df_state_abrev.dropna()
df_state_abrev.reset_index(drop=True, inplace=True)
```

```python
df_pop = df_pop.merge(df_state_abrev)
```

```python
df_pop.head()
```

```python
df_cohorts.head()
```

```python
df_cohorts = df_cohorts[["INSTNM", "STABBR"]]

mask_non_state_colleges = np.logical_not(df_cohorts["STABBR"]
                                                .isin(df_pop["Abbreviation"]))

df_cohorts.drop(df_cohorts[mask_non_state_colleges].index,
               inplace=True)

df_cohorts.reset_index(inplace=True, drop=True)
```

```python
school_count = df_cohorts.groupby("STABBR", as_index=False).count()
```

```python
school_count.rename(columns = {"INSTNM": "Schools"},
                    inplace=True)
```

```python
df_pop = df_pop.merge(school_count.rename(columns={"STABBR": "Abbreviation"}))

df_pop.head()
```

```python
df_pop["Population / School"] = df_pop["Population"] / df_pop["Schools"]
df_pop["Rank of Population / School"] = df_pop["Population / School"].rank()
```

```python
df_pop.head()
```

```python
df_uni_ranks.head()
```

```python
df_cohorts.head()
```

```python
df_cohorts["Score"] = np.nan
```

```python
not_in_ranked = ~df_cohorts["INSTNM"].isin(df_uni_ranks["Name"])
not_in_ranked.head(16)
```

```python
df_cohorts.loc[not_in_ranked, "Score"] = 1

df_cohorts["Score"].isnull().sum()

in_ranked = np.logical_not(not_in_ranked)
```

```python
names_ranked = df_cohorts.loc[in_ranked, "INSTNM"]
names_midlevel = df_uni_ranks.loc[df_uni_ranks["Rank"] > 100, "Name"]
mask_midlevel = df_cohorts["INSTNM"].isin(names_midlevel)
df_cohorts.loc[mask_midlevel, "Score"] = 3
```

```python
df_cohorts.loc[df_cohorts["Score"].isnull(), "Score"] = 5

df_cohorts.head()
```

```python
sum((df_cohorts["Score"] == 5)) + sum(df_cohorts["Score"] == 3)
```

```python
scores_by_state = df_cohorts.groupby("STABBR", as_index=False)["Score"].sum()

scores_by_state.sort_values("Score", ascending=False).head(10)
```

```python
df_pop = df_pop.merge(scores_by_state.rename(columns={"STABBR": "Abbreviation",
                                            "Score": "State Score"}))

df_pop.head(5)
```

```python
df_pop["Population / Score"] = df_pop["Population"] / df_pop["State Score"]
df_pop.head()
```

```python
df_pop["Rank of Population / Score"] = df_pop["Population / Score"].rank()
df_pop.head()
```

```python
df_pop = df_pop.sort_values(by="Rank of Population / Score")
df_pop
```
